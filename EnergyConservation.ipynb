```python
import solver as solv
import numpy as np
import matplotlib.pyplot as plt
```

Compute $\dot{n}/n$ and $\dot{E}/E$ as indicators of energy conservation

```python
# different bin sizes for momenta
ymin = 0.01
ymax = 50

ybin_0 = 10
ybin_1 = 20
ybin_2 = 30
ybin_3 = 40
ybin_4 = 50
ybin_5 = 100


# spans
x_span = (0.01,50)
y_span = (ymin,ymax)

# interaction strength
interaction_strength = 0.15
```

```python
sol_0 = solv.BE_solver(x_span,y_span,ybin_0,interaction_strength)
sol_1 = solv.BE_solver(x_span,y_span,ybin_1,interaction_strength)
sol_2 = solv.BE_solver(x_span,y_span,ybin_2,interaction_strength)
sol_3 = solv.BE_solver(x_span,y_span,ybin_3,interaction_strength)
sol_4 = solv.BE_solver(x_span,y_span,ybin_4,interaction_strength)
sol_5 = solv.BE_solver(x_span,y_span,ybin_5,interaction_strength)
```

```python
# get the time values
xs_0 = sol_0[0]
xs_1 = sol_1[0]
xs_2 = sol_2[0]
xs_3 = sol_3[0]
xs_4 = sol_4[0]
xs_5 = sol_5[0]

# get the distribution function values
fs_0 = sol_0[1]
fs_1 = sol_1[1]
fs_2 = sol_2[1]
fs_3 = sol_3[1]
fs_4 = sol_4[1]
fs_5 = sol_5[1]

# get momentum values
ys_0 = sol_0[2]
ys_1 = sol_1[2]
ys_2 = sol_2[2]
ys_3 = sol_3[2]
ys_4 = sol_4[2]
ys_5 = sol_5[2]
```

```python
f_0 = [fs_0[i][10] for i in range(len(ys_0))]
f_1 = [fs_1[i][10] for i in range(len(ys_1))]
f_2 = [fs_2[i][10] for i in range(len(ys_2))]
f_3 = [fs_3[i][10] for i in range(len(ys_3))]
f_4 = [fs_4[i][10] for i in range(len(ys_4))]
f_5 = [fs_5[i][10] for i in range(len(ys_5))]

# compute n_dot and E_dot for different times 
energy_conservation_0 = solv.energy_conservation(xs_0[10],ys_0,1,f_0)
energy_conservation_1 = solv.energy_conservation(xs_1[10],ys_1,1,f_1)
energy_conservation_2 = solv.energy_conservation(xs_2[10],ys_2,1,f_2)
energy_conservation_3 = solv.energy_conservation(xs_3[10],ys_3,1,f_3)
energy_conservation_4 = solv.energy_conservation(xs_4[10],ys_4,1,f_4)
energy_conservation_5 = solv.energy_conservation(xs_5[10],ys_5,1,f_5)

n_dot_0 = energy_conservation_0[0]
n_dot_1 = energy_conservation_1[0]
n_dot_2 = energy_conservation_2[0]
n_dot_3 = energy_conservation_3[0]
n_dot_4 = energy_conservation_4[0]
n_dot_5 = energy_conservation_5[0]

E_dot_0 = energy_conservation_0[1]
E_dot_1 = energy_conservation_1[1]
E_dot_2 = energy_conservation_2[1]
E_dot_3 = energy_conservation_3[1]
E_dot_4 = energy_conservation_4[1]
E_dot_5 = energy_conservation_5[1]
```

```python
print(n_dot_0)
print(n_dot_1)
print(n_dot_2)
print(n_dot_3)
print(n_dot_4)
print(n_dot_5)

print("energy:")

print(E_dot_0)
print(E_dot_1)
print(E_dot_2)
print(E_dot_3)
print(E_dot_4)
print(E_dot_5)
```
