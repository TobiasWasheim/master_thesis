```python
import matplotlib.pyplot as plt
from solver import solver
from collisions import test_function, test_solution, f0
```

```python
x_start, x_end = 0, 5
y_start, y_end = 0, 1
y_gridpoints = 100
Gamma = 1

x_span = (x_start,x_end)
y_span = (y_start,y_end,y_gridpoints)


solv = solver(x_span,y_span,Gamma,f0,test_function)

xs = solv[0]
ys = solv[1]    
fs = solv[2]
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 10))
N = [[25,50],[75,-1]]


for i in range(0,2):
    for j in range(0,2):
        n = N[i][j]
        sol = [test_solution(ys[n],x) for x in xs]
        axs[i,j].plot(xs,fs[n,:], label="Numerical Solution")
        axs[i,j].plot(xs,sol,ls="dashed", label="Analytical Solution")
        axs[i,j].set_xlabel("x")
        axs[i,j].set_ylabel("f(x,y)")
        axs[i,j].set_title(f"y={ys[n]:0.2f}")
        axs[i,j].legend()
plt.suptitle("Comparison between analytical and numerical solution with constant y")
fig.tight_layout()

N = [[10,25],[35,-1]]
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 10))
for i in range(0,2):
    for j in range(0,2):
        n = N[i][j]
        sol = [test_solution(y,xs[n]) for y in ys]
        axs[i,j].plot(ys,fs[:,n], label="Numerical Solution")
        axs[i,j].plot(ys,sol,ls="dashed", label="Analytical Solution")
        axs[i,j].set_xlabel("y")
        axs[i,j].set_ylabel("f(x,y)")
        axs[i,j].set_title(f"x={xs[n]:0.2f}")
        axs[i,j].legend()
plt.suptitle("Comparison between analytical and numerical solution with constant x")
fig.tight_layout()
```
